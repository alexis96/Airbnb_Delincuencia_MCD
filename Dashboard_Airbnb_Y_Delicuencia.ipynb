# Airbnb vs Delicuencia

## Column 1

### Distribución de airbnb

Podemos observar el mapa de densidad de airbnb's en la ciudad de México y a la derecha dos gráficos de totales, en la que apreciamos la relación entre el total de airbnb’s  y robos por alcaldía

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

df = pd.read_csv("./data/data_result.csv")
df = df.loc[df.last_review > "2020-01-01"]
fig = px.density_mapbox(df, lat='latitude', lon='longitude', hover_data=["name","alcaldia"] ,radius=2,
                        center=dict(lat=19.36, lon=-99.13), zoom=10,
                        mapbox_style="stamen-terrain")
fig.show()
```

## Column 2

### Total airbnb por alcaldia

```python
# Creamos el dataframe para facilitar la graficacion
df_bar_charts = df.groupby(["alcaldia"]).agg({"Robo a transeúnte en vía pública":"max", "name":"count"}).reset_index()
df_bar_charts = df_bar_charts.rename(columns={'name':'total_airbnb', "Robo a transeúnte en vía pública":"total_robos" })
df_bar_charts = df_bar_charts.sort_values(by="total_airbnb",ascending=False)
margin = go.layout.Margin(l=20, r=20, b=20, t=30)

fig = px.bar(df_bar_charts, x="alcaldia", y="total_airbnb",)
fig.update_layout(margin=margin)
fig.show()
```

### Robo a transeute por alcaldia

```python
fig = px.bar(df_bar_charts, x="alcaldia", y="total_robos",)
fig.update_layout(margin=margin)
fig.show()
```

# Extras

## Column 1

### Acumulado por alcaldia y tipo de cuarto

Total de airbnbs por alcaldia y tipo de cuarto, donde vemos que los de tipo departamento predominan ya que es el atractivo mas grande y versatil que ofrece la aplicación.

```python
df_bar_charts_acum= df.groupby(["alcaldia","room_type"]).agg({"name":"count"}).reset_index()
df_bar_charts_acum = df_bar_charts_acum.rename(columns={'name':'total', 'room_type':'tipo_cuarto'})
df_bar_charts_acum = df_bar_charts_acum.sort_values(by='total', ascending=False)
fig = px.bar(df_bar_charts_acum, x="alcaldia", y="total", color="tipo_cuarto")
fig.show()
```

## Column 2

### Acumulado de delitos por robo y alcaldia

En el acumulado de los distintos tipos de robo, podemos ver cierta relación con las alcaldias con mayor concentración de airbnb's a excepción de Iztapalapa que se posiciona como la número uno en robos.

```python
df_bar_charts_acum_del= df.groupby(["alcaldia"]).aggregate("max").reset_index()
df_bar_charts_acum_del = df_bar_charts_acum_del.sort_values(by="Robo a transeúnte en vía pública", ascending=False)
fig = px.bar(df_bar_charts_acum_del, x="alcaldia", y=["Robo a conductor/pasajero de vehículo", "Robo a pasajero a bordo de microbús", "Robo a pasajero a bordo de taxi","Robo a pasajero a bordo del Metro","Robo a transeúnte en vía pública"])
fig.show()
```
